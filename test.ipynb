```python
def quadrato(a:int, b:int):
    '''questa funzione calcola l'area e il perimetro del quadrato'''

    if a > 0 and b > 0:
        risultatoArea = a * b
        risultatoPerimetro = a * 4
        return risultatoArea, risultatoPerimetro
    else:
        print('devono essere positivi')

```

```python
area, perimetro = quadrato(6, 6)
print(area)
```
